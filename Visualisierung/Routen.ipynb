#### Load trajectories

```python
traj_path = "../StrassenNetz/new_edges_4326.geojson"
perimeter_city_path = "../WorldMove/output/City_Polygon.json"

with open(traj_path, "r") as f:
    traj_open = json.load(f)

with open(perimeter_city_path, "r") as f2:
    perimeter_city_open = json.load(f2)
```

```python
data = gpd.read_file(traj_path)

bridges_tunnels = data[
    (data["bridge"].fillna("").astype(str) == "yes") |
    (data["tunnel"].notna())
]

bridges_tunnels.to_file("../StrassenNetz/bridges_tunnels.geojson", driver="GeoJSON")
```
